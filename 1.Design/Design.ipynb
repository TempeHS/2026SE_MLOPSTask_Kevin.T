# Design

## Defining the business problem to be solved

NASCAR teams would like an additional method to predict where a selected driver would finish in a specific race to help either the team to see specifically where their driver is excelling or not, or identify possible other racers of concern. This system could also be provided to fans to see where their favourite driver is likely to end up, as currently due to the nature of NASCAR races it was quite hard to know. To fix these issues, NASCAR would like you to develop a finishing position prediction service for teams, drivers and fans based on multiple years of past race records.

## Refactoring the business problem into a machine learning problem

Create a NASCAR finishing prediction service based on data from past race records that is able to be used by fans and teams

## Researching available data.

The data that I have gathered is a record of all of the NASCAR races from 2017-2024, and it was sourced from Kaggle.

- The data columns are:

| Column | Data |
| ------ | -----|
| year | Year of the race |
| race_num | The order in the season that the race took place |
| track | The racetrack used |
| track_type | The type of track e.g. superspeedway, speedway, short track, road course |
| fin | Finishing position
| start | Starting/qualifying position
| car_num | The number that the driver has on their car |
| driver | The driver of the car |
| manu | The manufacturer of the car (Toyota, Chevrolet or Ford) |
| team_name | The name of the driver's team |
| laps | How many laps the driver completed in the race |
| laps_led | How many laps the driver led the field in the race |
| status | The status of the car (if they were running or in an accident and DNFed) |
| points | How many points the driver got from the race |
| stage_1 | Finishing position in stage 1 |
| stage_2 | Finishing position in stage 2 |
| stage_3_or_duel | Finishing position in stage 3 (only in coca cola 600 or the duel) |
| stage_points | Total stage points earned in the race |

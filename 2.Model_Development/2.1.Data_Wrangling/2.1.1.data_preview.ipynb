#### Graphically present the data

At this early stage you are just wanting to evaluate the data, the below simple plots lets you look the data in different ways to inform your model design and data wrangling approaches.

```python
# plot a line graph 
data_frame.plot(lw=1)
```

```python
# Plot a histogram of a column
plt.hist(data_frame['points'], bins=300)
plt.title(f"Histogram of {data_frame['points'].name}")
plt.ylabel('Count')
plt.xlabel(f'{data_frame["points"].name}')
plt.show()
```

```python
# Scatter plot 2 columns to see the relationship
plt.scatter(data_frame["fin"], data_frame["points"], s=1)
plt.title(f"Scatter of {data_frame['fin'].name} against {data_frame['points'].name}")
plt.ylabel(f"{data_frame['points'].name} Data")
plt.xlabel(f"{data_frame['fin'].name} Data")
plt.show()
```

```python
# Scatter plot multiples columns to see the relationship
x_plot = ['start', 'laps_led']
for col in x_plot:
    plt.scatter(data_frame[col], data_frame['fin'], marker='x')
plt.title(f"Scatter of {*x_plot,} against {data_frame['fin'].name}")
plt.ylabel(f"{data_frame['fin'].name} Data")
plt.xlabel(f"{*x_plot,} Data")
plt.show()
```

```python
# Scatter plot 2 columns in separate charts with a shared y-axis
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True)
plt.suptitle(f"Scatter of {data_frame['start'].name} and {data_frame['laps_led'].name} against {data_frame['fin'].name}")
ax1.set_ylabel(f"{data_frame['fin'].name} Data")

ax1.scatter(data_frame["start"], data_frame["fin"])
ax1.set_xlabel(f"{data_frame['start'].name} Data")

ax2.scatter(data_frame["laps_led"], data_frame["fin"])
ax2.set_xlabel(f"{data_frame['laps_led'].name} Data")

plt.show()
```

```python
# 3D Scatter plot 3 columns to see the relationship

x_plot = ['BMI', 'BP']

fig = plt.figure()
plt.suptitle(f"3D Scatter of {*x_plot,} against {data_frame['Target'].name}")
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data_frame[x_plot[0]], data_frame[x_plot[1]], data_frame['Target'], color='blue')

x1_range = np.linspace(data_frame[x_plot[0]].min(), data_frame[x_plot[0]].max())
x2_range = np.linspace(data_frame[x_plot[1]].min(), data_frame[x_plot[1]].max())
X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)


ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')

plt.show()
```
